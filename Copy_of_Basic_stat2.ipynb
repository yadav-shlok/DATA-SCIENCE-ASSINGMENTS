```python
import numpy as np
import pandas as pd
from scipy import stats
import os
```

```python
print("Confidence Interval Analysis")
print()
```

```python
# Print-head durability data from the assignment (15 samples in millions of characters)
durability = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
n = len(durability)
sample_mean = durability.mean()
sample_std = durability.std(ddof=1)  # sample standard deviation
alpha = 0.01  # 99% confidence level
```

```python
print("Part 1: Print-head Durability Analysis")
print(f"Sample size: {n}")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample std dev: {sample_std:.4f}")
print()
```

```python
# (a) 99% CI using t-distribution (sample standard deviation)
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)
standard_error_t = sample_std / np.sqrt(n)
margin_error_t = t_critical * standard_error_t
```

```python
print("(a) 99% Confidence Interval using sample standard deviation (t-distribution)")
print(f"t-critical value (df={df}): {t_critical:.4f}")
print(f"99% CI: ({sample_mean - margin_error_t:.4f}, {sample_mean + margin_error_t:.4f})")
print("Note: Used t-distribution because population SD is unknown and sample size is small")
print()
```

```python
# (b) 99% CI using z-distribution (known population SD = 0.2)
sigma = 0.2  # known population standard deviation
z_critical = stats.norm.ppf(1 - alpha/2)
standard_error_z = sigma / np.sqrt(n)
margin_error_z = z_critical * standard_error_z
```

```python
print("(b) 99% Confidence Interval using known population SD = 0.2 (z-distribution)")
print(f"z-critical value: {z_critical:.4f}")
print(f"99% CI: ({sample_mean - margin_error_z:.4f}, {sample_mean + margin_error_z:.4f})")
print("Note: Used z-distribution because population SD is known")
print()
```

```python
csv_path = "/content/sample_data/sales_data_with_discounts.csv"
```

```python
# Try to load CSV, use sample data if permission denied
if os.path.exists(csv_path):
    try:
        sales_df = pd.read_csv(csv_path)
        net_sales = pd.to_numeric(sales_df["Net Sales Value"], errors='coerce').dropna()
        print(f"CSV loaded successfully. Sample size: {len(net_sales):,}")
    except:
        print("Could not access CSV file, using sample data")
        net_sales = np.array([115346.50, 89323.90, 102042.09, 112235.93, 19927.05])
else:
    print("CSV file not found, using sample data")
    net_sales = np.array([115346.50, 89323.90, 102042.09, 112235.93, 19927.05])
```

```python
# Calculate 99%CI for Net Sales Value
n_sales = len(net_sales)
df_sales = n_sales - 1
t_critical_sales = stats.t.ppf(1 - alpha/2, df_sales)
standard_error_sales = net_sales.std(ddof=1) / np.sqrt(n_sales)
margin_error_sales = t_critical_sales * standard_error_sales
```

```python
print()
print("99% Confidence Interval for Net Sales Value (t-distribution)")
print(f"Sample size: {n_sales}")
print(f"Sample mean: Rs {net_sales.mean():,.0f}")
print(f"99% CI: Rs ({net_sales.mean() - margin_error_sales:,.0f}, Rs {net_sales.mean() + margin_error_sales:,.0f})")
```
