```python
# Exercise 1: Prime Number Checker
def is_prime(number):
    """Check if a number is prime. Returns True if prime, False otherwise."""
    if number <= 1:
        return False
    if number == 2:
        return True

    # Check for divisibility from 2 to square root of number
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True
```

```python
# Test cases
test_numbers = [17, 15, 2, 1, 29, 100]
print("Exercise 1: Prime Number Checker")
for num in test_numbers:
    result = is_prime(num)
    print(f"{num}: {'Prime' if result else 'Not prime'}")
print()
```

```python
# Exercise 4: Word Counter
def count_words(text):
    """Count frequency of each word in given text."""
    # Clean text: lowercase, remove punctuation
    words = text.lower().replace('.', '').replace(',', '').split()

    word_count = {}
    for word in words:
        if word:  # Skip empty strings
            word_count[word] = word_count.get(word, 0) + 1

    return word_count
```

```python
# Assignment example
sample_text = "This is a sample text. This text will be used to demonstrate the word counter."
print("Exercise 4: Word Counter")
print(f"Input: {sample_text}")
```

```python
word_counts = count_words(sample_text)
print("Output:")
for word, count in sorted(word_counts.items()):
    print(f"  {word}: {count}")
print()
```

```python
# Exercise 5: Palindrome Checker
def is_palindrome(text):
    """Check if string is palindrome (ignores spaces, punctuation, case)."""
    # Keep only alphabetic characters, lowercase
    cleaned = ''.join(char.lower() for char in text if char.isalpha())
    return cleaned == cleaned[::-1]
```

```python
# Test cases from assignment + extras
test_cases = ["racecar", "A man a plan a canal Panama", "hello", "12321", " madam "]
print("Exercise 5: Palindrome Checker")
for test in test_cases:
    result = is_palindrome(test)
    print(f"'{test}' -> {result}")
print()
```
